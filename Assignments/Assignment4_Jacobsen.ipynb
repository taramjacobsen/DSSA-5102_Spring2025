All we know at the start of our investigation is the <b>murder</b> occurred on <b>Jan.15, 2018</b> and that it took place in <b>SQL City<b>.

```python
#Write a query in SQL to view the crime scene report 
#for the murder occurred on Jan.15, 2018 and that it took place in SQL City.
#Select ALL records FROM the table WHERE...

pd.set_option('display.max_colwidth', None) # No max column width so I can read all of the description

with engine.connect() as connection: # Establish connection
    # Define query to select all records from crime scene report where date is 20180115 and city is SQL City and type is murder
    query = text(""" SELECT * 
                     FROM crime_scene_report
                     WHERE date == 20180115 AND city == 'SQL City' AND type == 'murder'
                     """)
    report = pd.read_sql(query, connection) # Use pd to read query with connection to db

report # Print result



```

Another clue! It looks like we have two witnesses! Now who are they and how do we find them?

```python
#Write two queries in SQL to find each witness in the person table. 
#For the first query as first_witness, select all the records from person where the address_street_name is Northwestern Drive and order by adress_number
#Fetch the first_witness

with engine.connect() as connection: # Establish connection
    # Define query to select all records from person where the address street name is Northwestern Dr
    # and order by address_number DESC so the last house appears first. Limit to first 5 rows.
    query = text(""" SELECT * 
                     FROM person
                     WHERE address_street_name == 'Northwestern Dr' 
                     ORDER BY address_number DESC
                     LIMIT 5
                     """)
    witness1 = pd.read_sql(query, connection) # Use pd to read query with connection to db

witness1 # Print result

# 14887, Morty Schapiro
```

We have identified our first witness as Morty Schapiro (14887). Now lets find the second!

```python
#For the third query as second_witness,
#Select all the records from person where the address_street_name is Frnaklin Ave and the name includes Annabel. 
#Fetch the second_witness

with engine.connect() as connection: # Establish connection
    # Define query to select all records from person where the address street name is Franklin Ave
    # and name includes Annabel
    query = text(""" SELECT * 
                     FROM person
                     WHERE address_street_name == 'Franklin Ave'
                     AND name LIKE '%Annabel%'
                     """)
    witness2 = pd.read_sql(query, connection) # Use pd to read query with connection to db

witness2 # Print result

# 16371, Annabel Miller
```

Three different people match this description! Now which one attended the SQL Symphony Concert three times? 

```python
#Let's find the killer!
#Remember to write our your query in the comments!


with engine.connect() as connection: # Establish connection
    # Define query to select all records from facebook_event_checkin where the event
    # is SQL Symphony Concert where they attended 3+ times and the date includes
    # 2017 December and the person_id is one of the above 
    query = text(""" SELECT * 
                     FROM facebook_event_checkin
                     WHERE event_name = 'SQL Symphony Concert'
                     GROUP BY person_id
                     HAVING COUNT(*) >= 3
                     AND date LIKE '%201712%'
                     AND person_id IN (78881, 90700, 99716)
                     """)
    clue5 = pd.read_sql(query, connection) # Use pd to read query with connection to db

clue5 # Print result

# Only one! 99716, Miranda Priestly
```
