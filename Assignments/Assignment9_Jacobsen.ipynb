#### CASE Statements
Case statements are similar to the if-then-else statements in programming. The data that meets the case statements in the database will be returned. You start a CASE statement with the word CASE and conclude it with an END. Between those keywords, you specify each condition with a WHEN [condition] THEN [value], where the condition and the values are specified by you.

Nield, Thomas. Getting Started with SQL (p. 71). O'Reilly Media. Kindle Edition. <br>

<br>
**Example Question:** What planets have similar orbital periods to Earth? 
<br>
**What tables are we joining? (If any):** planet_identifiers and orbital_info

```python
with engine.connect() as connection:  # Establish a connection
    query = text("""SELECT
                            planet_identifiers.planet_name AS name, 
                            orbital_info.orbital_period_days AS orbital_period,
                        CASE
                            WHEN orbital_info.orbital_period_days BETWEEN 0.8 AND 1.2
                                THEN 'Similar to Earth orbital period'
                                ELSE 'Not similar to Earth orbital period'
                            END AS orbit_classification
                     FROM planet_identifiers
                     RIGHT JOIN orbital_info
                     ON planet_identifiers.id = orbital_info.properties_id;

                                 """) #  SELECT planet name and orbital period AS name and orbial_period FROM JOINed tables planet_identifiers and orbital_info.
    # WHEN orbital_period is between 0.8 and 1.2, it is labeled 'Similar to Earth'
    # If not, it is labeled 'Not similar to Earth'. END AS labels this column as orbital_classification. 
    query2 = pd.read_sql(query, connection) 
    
# Print the results
query2
```

**Question:** What systems have a planet which is similar in temperature to Earth? 
<br>
**What tables are we joining? (If any):** planet_properties and insolation_and_temp

```python
with engine.connect() as connection:
    query4 = text("""
        SELECT
            planet_properties.planet_id,
            planet_properties.planet_radius_earth_radius,
            insolation_and_temp.equilibrium_temperature_k,
            CASE
                WHEN insolation_and_temp.equilibrium_temperature_k BETWEEN 175 AND 270
                     AND planet_properties.planet_radius_earth_radius BETWEEN 0.8 AND 1.2
                    THEN 'Potentially Habitable'
                ELSE 'Not Potentially Habitable'
            END AS potential_habitability_label
        FROM insolation_and_temp
        JOIN planet_properties
            ON insolation_and_temp.properties_id = planet_properties.properties_id;
    """) # SELECT planet id, radius, and equilibrium temp FROM JOINED insolation temp and planet properties tables via id. WHEN equilibrium temp is between 175-270 AND radius is between 0.8 and 1.2 Eart's, label
    # 'Potentially Habitable'. If not, label 'Not Potentially Habitable'. END AS potential_habitability_label for column name. 

    query4_result = pd.read_sql(query4, connection)

# Print the results
query4_result
```

```python
with engine.connect() as connection:
    query5 = text("""
        SELECT
            COUNT(*) AS missing_equilibrium_temp_count
        FROM insolation_and_temp
        WHERE equilibrium_temperature_k IS NULL;
    """) # SELECT equilibrium temp FROM insolation_and_temp. COUNT NULLs as missing_equilibrium_temp_count

    query5_result = pd.read_sql(query5, connection)

# Print the results
query5_result
```

**Question:** What are the max and min orbital ratios of planets in circumbinary systems?
<br>
**What table(s) are we joining? (If any)** orbital_ratios, system_info

```python
from sqlalchemy import text

with engine.connect() as connection:
    query7 = text("""
        SELECT 
            MAX(orats.ratio_semi_major_axis_to_stellar_radius) AS max_ratio,
            MIN(orats.ratio_semi_major_axis_to_stellar_radius) AS min_ratio
        FROM orbital_ratios AS orats
        JOIN system_info AS si ON orats.orbital_id = si.system_id
        WHERE si.is_circumbinary = 1;
    """) # SELECT orbital ratio FROM JOINed tables orbital_ratio AS orats and system info AS si WHERE is_circumbinary is TRUE. Find the MAX of orbital ratio and place in column max_ratio
    # Find the MIN and place in column min_ratio

    query7_result = pd.read_sql(query7, connection)

# Print the results
query7_result
```

Combine CASE statement with Min() and Max() for a more detailed query of your data:
<br><br>
**Question:**  Is there a difference if the system is not circumbinary? 
<br>
**What table(s) are we joining?** orbital_ratio, system_info

```python
with engine.connect() as connection:
    query8 = text("""
        SELECT 
            MAX(CASE WHEN si.is_circumbinary = 1 THEN orats.ratio_semi_major_axis_to_stellar_radius END) AS max_ratio_circumbinary,
            MAX(CASE WHEN si.is_circumbinary = 0 THEN orats.ratio_semi_major_axis_to_stellar_radius END) AS max_ratio_non_circumbinary,
            MIN(CASE WHEN si.is_circumbinary = 1 THEN orats.ratio_semi_major_axis_to_stellar_radius END) AS min_ratio_circumbinary,
            MIN(CASE WHEN si.is_circumbinary = 0 THEN orats.ratio_semi_major_axis_to_stellar_radius END) AS min_ratio_non_circumbinary
        FROM orbital_ratios AS orats
        JOIN system_info AS si ON orats.orbital_id = si.system_id;
    """) # SELECT orbital ratio FROM JOINed tables orbital_ratio AS orats and system info AS si. WHERE is_circumbinary is TRUE, find the MAX of orbital ratio and place in column max_ratio_circumbinary
    # and find the MIN of orbtial ratio and place in column min_ratio_circumbinary. Repeat for WHERE is_circumbunary is FALSE, placing results in columns max_ratio_non_circumbinary and min_ratio_non_circumbinary

    query8 = pd.read_sql(query8, connection)

# Print the results
query8
```

**Question:** What is the average orbital ratio of circumbinary versus noncirumbinary planets. 
<br>

```python

with engine.connect() as connection:
    query9 = text("""
        SELECT 
            AVG(CASE WHEN si.is_circumbinary = 1 THEN orats.ratio_semi_major_axis_to_stellar_radius END) AS avg_ratio_circumbinary,
            AVG(CASE WHEN si.is_circumbinary = 0 THEN orats.ratio_semi_major_axis_to_stellar_radius END) AS avg_ratio_non_circumbinary
        FROM orbital_ratios AS orats
        JOIN system_info AS si ON orats.orbital_id = si.system_id;
    """) # SELECT orbital ratio FROM JOINed tables orbital_ratio AS orats and system info AS si. Find the AVG of the orbital ratio when
    # system_info=1 and place in column avg_ratio_circumbinary. Repeat for is_circumbinary=0 and place in column avg_ratio_non_circumbinary.

    query9 = pd.read_sql(query9, connection)

# Print the results
query9
```
